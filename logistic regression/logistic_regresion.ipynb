```python
df = pd.read_csv('Social_Network_Ads.csv')
df
```

```python
df_new = df.drop(['User ID'], axis=1)
df_new
```

```python
X = df_new.iloc[:, :-1]
y = df_new.iloc[:, -1]
```

```python
# Split Data into Training and Testing Part
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0, stratify = y)
```

```python
# Importing OneHotEncoder and MinMaxScaler
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
```

```python
# Initialize OneHotEncoder
ohe = OneHotEncoder(drop = 'first', sparse = False)
X_train_gender = ohe.fit_transform(X_train[['Gender']])
X_test_gender = ohe.transform(X_test[['Gender']])
```

```python
# Append transformed gender with age and salary to make it as input data
X_train_t = np.hstack((X_train_gender, X_train.drop('Gender', axis=1)))
X_test_t = np.hstack((X_test_gender, X_test.drop('Gender', axis=1)))
```

```python
# Initialize and fitting scaler
sc = MinMaxScaler()
X_train_scaled = sc.fit_transform(X_train_t)
X_test_scaled = sc.transform(X_test_t)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X_train_scaled, y_train)
model.score(X_test_scaled, y_test)
```

```python
X_scaled = np.vstack((X_train_scaled, X_test_scaled))
```

```python
X_new = X.copy()
X_new['Gender'] = X_new['Gender'].replace({'Male':0, 'Female':1})
```

```python
X_new_scaled = sc.fit_transform(X_new)
```

```python
from sklearn.model_selection import cross_val_score
model = KNeighborsClassifier()
cv = cross_val_score(model, X_new_scaled, y, cv = 10)
```

```python
cv
```

```python
cv.mean()
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
param = [{
    'n_neighbors':[3, 5, 7, 9],
    'p':[1, 2, 3]
}]
```

```python
model_cv = GridSearchCV(model, param_grid=param, cv = 5)
model_cv.fit(X_new_scaled, y)
```

```python
print(model_cv.best_params_)
```

```python
print(model_cv.best_score_)
```

```python
model = KNeighborsClassifier(n_neighbors = 9, p = 3)
cv = cross_val_score(model, X_new_scaled, y, cv = 10)
cv
```

```python
cv.mean()
```

```python
model_cv = GridSearchCV(model, param_grid=param, cv = 5, verbose = 4)
model_cv.fit(X_new_scaled, y)
```

```python
model_cv = GridSearchCV(model, param_grid=param, cv = 5, n_jobs = -1)
model_cv.fit(X_new_scaled, y)
```

```python
model_cv = RandomizedSearchCV(model, param_distributions=param, cv = 5, n_iter = 3, verbose = 4)
model_cv.fit(X_new_scaled, y)
```

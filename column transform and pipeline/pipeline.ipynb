```python
# Binding Everything together
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
```

```python
from google.colab import files
files.upload()
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
df.head(2)
```

```python
# Dropping User ID
df = df.drop(('User ID'), axis=1)
```

```python
# Separating Input and Output
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
# Splitting data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 0, stratify = y)
```

```python
# Initializing transformers for the input to Pipeline
transformer_1 = ColumnTransformer(
    [
        ('One Hot Cat Encoding', OneHotEncoder(drop = 'first', sparse = False), [0]),
        ('Age Scaling', MinMaxScaler(), [1]),
        ('Filling Null with Mean', SimpleImputer(), [2])
    ], remainder = 'passthrough'
)

transformer_2 = ColumnTransformer(
    [
        ('Salary Scaling', MinMaxScaler(), [2])
    ], remainder = 'passthrough'
)

model = KNeighborsClassifier()
```

```python
# Creating Pipeline
pipe = Pipeline([
    ('Transformer 1', transformer_1),
    ('Transformer 2', transformer_2)
])
```

```python
# Fitting pipe to train data
X_train_transformed = pipe.fit_transform(X_train)
model.fit(X_train_transformed, y_train)
```

```python
# Getting Prediction on Test Data
X_test_transformed = pipe.transform(X_test)
model.score(X_test_transformed, y_test)
```

```python
# Applying Pipeline with Cross Validation
# Creating Pipeline
pipe = Pipeline([
    ('Transformer 1', transformer_1),
    ('Transformer 2', transformer_2),
    ('Model', model)
])
cv_scores = cross_val_score(pipe, X_train, y_train, cv = 10, n_jobs = -1)
```

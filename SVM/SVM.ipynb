```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from google.colab import files
files.upload()
```

```python
df = pd.read_csv("Social_Network_Ads.csv")
df.head(2)
```

```python
df = df.drop(['User ID'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:-1], df.iloc[:, -1], test_size = 0.2, random_state = 0, stratify = df.iloc[:, -1])
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
model.fit(X_train, y_train)
```

```python
model.get_params()
```

```python
model.score(X_train, y_train), model.score(X_test, y_test)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, :-1], df.iloc[:, -1], test_size = 0.2, random_state = 0, stratify = df.iloc[:, -1])
```

```python
X_train
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first', sparse = False, handle_unknown='ignore')
X_train_gender = ohe.fit_transform(X_train[['Gender']])
X_test_gender = ohe.transform(X_test[['Gender']])
```

```python
X_train_new = np.hstack((X_train_gender, X_train.iloc[:, 1:]))
X_test_new = np.hstack((X_test_gender, X_test.iloc[:, 1:]))
```

```python
model = SVC()
model.fit(X_train_new, y_train)
```

```python
model.score(X_train_new, y_train), model.score(X_test_new, y_test)
```

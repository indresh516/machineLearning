```python
import numpy as np
import pandas as pd
```

```python
n = 10

g = ['M', 'F', 'T']
g = [g[np.random.randint(0, 3, 1)[0]] for i in range(n)]

ed = ['10th', '12th', 'diploma', 'ug', 'pg', 'doctorate']
ed = [ed[np.random.randint(0, 6, 1)[0]] for i in range(n)]

r = ['pass', 'fail']
r = [r[np.random.randint(0, 2, 1)[0]] for i in range(n)]

age = np.random.randint(10, 30, n)
```

```python
df =  pd.DataFrame({
    'gender' : g,
    'education': ed,
    'age' : age,
    'result': r
})
```

```python
df_v = pd.read_csv("https://raw.githubusercontent.com/edyoda/data-science-complete-tutorial/master/Data/house_rental_data.csv.txt")
```

```python
df_v
```

```python
import matplotlib.pyplot as plt
```

```python
df_v_gp1 = df_v.groupby(['TotalFloor'], as_index=False)['Price'].mean()
```

```python
df_v_gp1
```

```python
plt.figure(figsize= (20, 5))
plt.bar(df_v_gp1['TotalFloor'], df_v_gp1['Price'])
```

```python
df_v_gp1 = df_v.groupby(['Floor'], as_index=False)['Price'].mean()
plt.figure(figsize= (20, 5))
plt.bar(df_v_gp1['Floor'], df_v_gp1['Price'])
```

```python
df_v_gp1 = df_v.groupby(['Floor', 'Bedroom'], as_index=False)['Price'].mean()
plt.figure(figsize= (20, 5))
plt.bar(df_v_gp1['Floor'].astype(str)+'_'+df_v_gp1['Bedroom'].astype(str), df_v_gp1['Price'])
plt.xticks(rotation=90)
```
